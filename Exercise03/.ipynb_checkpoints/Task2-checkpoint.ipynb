```python
import pandas as pd
```

```python
data3 = pd.read_csv('data_DMAP_PCA_vadere.txt', sep=' ')

eigenvalues, eigenvectors = diffusion_map(data3, 7)
```

```python
def plot_pedestrian_path_3d(first_eigen_number: int =1):
    fig = plt.figure(figsize=(8, 8))
   
    ax = fig.add_subplot(111, projection='3d')  # Create a 3D axis

    ax.scatter(eigenvectors[:, first_eigen_number], eigenvectors[:, first_eigen_number+1], eigenvectors[:, first_eigen_number+2], c=range(len(eigenvectors)), cmap='viridis')

    
    ax.set_title(rf"$_{first_eigen_number}$, $_{first_eigen_number+1}$, $_{first_eigen_number+2}$ Eigenfunctions")
    ax.set_xlabel(rf"$_{first_eigen_number}$")
    ax.set_ylabel(rf"$_{first_eigen_number+1}$")
    ax.set_zlabel(rf"""          
                  
                  $_{first_eigen_number+2}$


                    """)
    ax.grid(True),


    plt.show()

for i in range(6):
    plot_pedestrian_path_3d(i)
```

```python
def plot_pedestrian_path_2d(first_eigen_number: int = 1):
    fig, ax = plt.subplots(figsize=(4, 4))

    ax.scatter(eigenvectors[:, first_eigen_number], eigenvectors[:, first_eigen_number + 1], c=range(len(eigenvectors)), cmap='viridis')

    ax.set_title(rf"$_{first_eigen_number}$, $_{first_eigen_number+1}$ Eigenfunctions")
    ax.set_xlabel(rf"$_{first_eigen_number}$")
    ax.set_ylabel(rf"$_{first_eigen_number+1}$")
    ax.grid(True)

    plt.show()

for i in range(6):
    plot_pedestrian_path_2d(i)
```
