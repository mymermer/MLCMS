## Task 1.3

```python
#-----3rd Part-----

# Load the data
data = np.loadtxt('../data/data_DMAP_PCA_vadere.txt')
```

```python
plt.plot(data[:,0],data[:,1], color='r', label='1st Pedestrian')
plt.plot(data[:,2],data[:,3], color='g', label='2nd Pedestrian')
plt.legend()
plt.savefig('images/ex3task1-3.1.png')
plt.show()
```

```python
# With 2 principal components
pca_energy, variance, centered_data, U, S, Vt, transformed_data, reconstructed_data = pca_def(data,2)
```

```python
plt.plot(reconstructed_data[:,0],reconstructed_data[:,1], color='r', label='1st Pedestrian')
plt.plot(reconstructed_data[:,2],reconstructed_data[:,3], color='g', label='2nd Pedestrian')
plt.title("Reconstruced data using 2 Principal Components")
plt.legend()
plt.savefig('images/ex3task1-3.2.png')
plt.show()
print("PCA Energy",pca_energy)
```

```python
# With 3 principal components
pca_energy, variance, centered_data, U, S, Vt, transformed_data, reconstructed_data = pca_def(data,3)
```

```python
plt.plot(reconstructed_data[:,0],reconstructed_data[:,1], color='r', label='1st Pedestrian')
plt.plot(reconstructed_data[:,2],reconstructed_data[:,3], color='g', label='2nd Pedestrian')
plt.title("Reconstruced data using 3 Principal Components")
plt.legend()
plt.savefig('images/ex3task1-3.3.png')
plt.show()
print("PCA Energy",pca_energy)
```
