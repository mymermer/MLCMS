```python
from gensim.models import KeyedVectors
from UMAP_GPU_util import UMAPVisualizer
```

## Dimensionality Reduction on Swiss Roll dataset (10k points)

```python
swiss10k = UMAPVisualizer(embedding_n_neighbors=100, pairwise_distances_sample_size=200) #intilize the class
```

```python
swiss10k.swiss(n_samples=10000) #umap on swiss roll 
print("Elapsed Time: ",swiss10k.elapsed_time)
```

```python
swiss10k.plot_swiss() #plot the orginal swiss roll
swiss10k.plot() #plot the umap of swiss roll
```

```python
print("trustworthiness: ",swiss10k.tw)
```

## Dimensionality Reduction on Google news word Embeddings

```python
# Load the Word2Vec model
binary_file_path = './GoogleNews-vectors-negative300.bin' #Google's pre-trained Word2Vec model
word2vec_model = KeyedVectors.load_word2vec_format(binary_file_path, binary=True) #load the model
```

### 10K Points

```python
word2vec10k = UMAPVisualizer(embedding_n_neighbors=10, pairwise_distances_sample_size=200)  #intilize the class
```

```python
word2vec10k.word2vec(word2vec_model, subset_size=10000) #umap on word2vec
print(f"Elapsed time: {word2vec10k.elapsed_time} s")
```

```python
word2vec10k.plot() #plot the umap of word2vec
```

```python
print("trustworthiness: ",word2vec10k.tw)
```
