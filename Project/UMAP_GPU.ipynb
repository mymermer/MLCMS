```python
from gensim.models import KeyedVectors
from UMAP_GPU_util import UMAPVisualizer
```

## Dimensionality Reduction on Swiss Roll dataset (10k points)

```python
swiss10k = UMAPVisualizer(embedding_n_neighbors=100, pairwise_distances_sample_size=200)
```

```python
swiss10k.swiss(n_samples=10000) #10k points
print("Elapsed Time: ",swiss10k.elapsed_time)
```

```python
swiss10k.plot_swiss()
swiss10k.plot()
```

```python
print("trustworthiness: ",swiss10k.tw)
```

## Dimensionality Reduction on Google news word Embeddings

```python
# Load the Word2Vec model
binary_file_path = './GoogleNews-vectors-negative300.bin' #download it and put it to your directory
word2vec_model = KeyedVectors.load_word2vec_format(binary_file_path, binary=True)
```

### 10K Points

```python
word2vec10k = UMAPVisualizer(embedding_n_neighbors=10, pairwise_distances_sample_size=200)
```

```python
word2vec10k.word2vec(word2vec_model, subset_size=10000)
print(f"Elapsed time: {word2vec10k.elapsed_time} s")
```

```python
word2vec10k.plot()
```

```python
print("trustworthiness: ",word2vec10k.tw)
```
