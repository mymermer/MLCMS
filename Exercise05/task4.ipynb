```python
from utils_task4 import *
from scipy.integrate import solve_ivp
import numpy as np
```

```python
file_path = "./data/takens_1.txt"

line_numbers = []
data_from_first_column = []

with open(file_path, "r") as file:
    for line_number, line in enumerate(file, start=0):
        # Split the line into columns
        columns = line.strip().split()

        # Extract data from the first column
        first_column_data = float(columns[0])

        # Append line number and data to lists
        line_numbers.append(line_number)
        data_from_first_column.append(first_column_data)

# Plot the data 
plt.scatter(line_numbers, data_from_first_column)
plt.title("Takens' first map")
plt.xlabel("Line number")
plt.ylabel("Data from first column")
plt.show()
```

```python
plt.scatter(data_from_first_column, data_from_first_column[5:] + data_from_first_column[:5])
plt.title("Takens' first map")
plt.xlabel("Line number")
plt.ylabel("Data from first column")
plt.show()
```

```python
#Initial Parameters
sigma = 10
rho = 28
beta = 8/3

# Initial state
initial_state = [10, 10, 10]

# time points with time step 0.01
t = np.arange(0, 100, 0.01)

# Solve the Lorenz system
solution = solve_ivp(lorenz_system, [t.min(), t.max()], initial_state, args=(sigma, rho, beta), t_eval=t)

coordinates=solution.y[0:3].tolist()


# Plotting the Lorez Attractor
plot_lorenz_attractor(solution)
```

```python
x2=coordinates[0][5:]+ coordinates[0][:5]
x3= coordinates[0][10:]+ coordinates[0][:10]



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(coordinates[0], x2,x3,lw=0.5)
ax.set_title("Lorenz Attractor")
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X3")
ax.legend()
plt.show()
```
