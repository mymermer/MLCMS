part 1

analyze the map by varying the parameter r from 0 to 2 

from 2 to 4, noting bifurcations and steady states. 

A bifurcation diagram for  r between 0 and 4

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_map(r, x):
    return r * x * (1 - x)
```

```python
r_values1 = np.linspace(0, 2, 500)  
x_values1 = []
r_plot1 = []

# Iterate 
for r in r_values1:
    x = 0.1
    for _ in range(200):  
        x = logistic_map(r, x)
    for _ in range(20):  
        x = logistic_map(r, x)
        x_values1.append(x)
        r_plot1.append(r)

# Plot 
plt.figure(figsize=(12, 8))
plt.plot(r_plot1, x_values1, ',k', alpha=0.5)
plt.title('Bifurcation Diagram of the Logistic Map (r from 0 to 2)')
plt.xlabel('r')
plt.ylabel('x')
plt.show()
```

```python

r_values2 = np.linspace(2, 4, 500)  # Adjusted range of r values

# Reinitialize arrays for plotting
x_values2 = []
r_plot2 = []

# Iterate with fewer steps and store results for plotting
for r in r_values2:
    x = 0.1
    for _ in range(200):  # Reduced number of initial iterations
        x = logistic_map(r, x)
    for _ in range(20):  # Reduced number of plotted iterations
        x = logistic_map(r, x)
        x_values2.append(x)
        r_plot2.append(r)

# Plot the bifurcation diagram
plt.figure(figsize=(12, 8))
plt.plot(r_plot2, x_values2, ',k', alpha=0.5)
plt.title('Bifurcation Diagram of the Logistic Map (r from 2 to 4)')
plt.xlabel('r')
plt.ylabel('x')
plt.show()
```

```python
# Values of r from 0 to 4
r_values3 = np.linspace(0, 4, 1000)  # Expanded range of r values

# Reinitialize arrays for plotting
x_values3 = []
r_plot3 = []

# Iterate with fewer steps and store results for plotting
for r in r_values3:
    x = 0.1
    for _ in range(200):  # Reduced number of initial iterations
        x = logistic_map(r, x)
    for _ in range(20):  # Reduced number of plotted iterations
        x = logistic_map(r, x)
        x_values3.append(x)
        r_plot3.append(r)

# Plot the bifurcation diagram
plt.figure(figsize=(12, 8))
plt.plot(r_plot2, x_values2, ',k', alpha=0.5)
plt.title('Bifurcation Diagram of the Logistic Map (r from 0 to 4)')
plt.xlabel('r')
plt.ylabel('x')
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the Lorenz system
def lorenz_system(t, state, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]
```

```python
# Parameters
sigma1 = 10
rho1 = 28
beta1 = 8/3

# Initial condition
initial_state1 = [10, 10, 10]

# Time points
t = np.linspace(0, 50, 10000)

# Solve the Lorenz system
solution = solve_ivp(lorenz_system, [t.min(), t.max()], initial_state1, args=(sigma1, rho1, beta1), t_eval=t)
```

```python
# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(solution.y[0], solution.y[1], solution.y[2])
ax.set_title("Lorenz Attractor")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
plt.show()
```
